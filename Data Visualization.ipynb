```python
import numpy as np
import pandas as pd

# Import the Altair library
import altair as alt
```

```python
iris = pd.read_csv('/content/drive/My Drive/HDS/Datasets/iris.csv')
iris.head()
```

```python
alt.Chart(iris).mark_point().encode(
    x= "sepal_length",
    y = "sepal_width"
    )
```

```python
alt.Chart(iris).mark_point().encode(
  x= alt.X("sepal_width" , title= "CANT"),
  y="sepal_length",
  color = "species"
).properties(title= "Iris Scatterplot of Sepal_length by Sepal_width by species")
```

```python
alt.Chart(iris).mark_point().encode(
    x=alt.X("sepal_width", scale=alt.Scale(zero=False)),
    y=alt.Y("sepal_length", scale=alt.Scale(zero=False)),
    color= "species"
).properties(title= "Sepal length vs Sepal Width by Species")
```

```python
alt.Chart(iris).mark_point().encode(
    x= alt.X("sepal_length",title="SEPAL LENGTH" , scale=alt.Scale(zero=False)),
    y= alt.Y("sepal_width", title= "SEPAL WIDTH" , scale= alt.Scale(zero=False)),
    color= alt.Color("species", title= "SPECIES")
).properties(title= "sepal width vs sepal length by species")
```

Line Plots using Altair

```python
x= np.arange(-10,11)
print("X:", x)

squared = x**2
print("Squared: ", squared)

powers = pd.DataFrame({'X': x, 'Squared' : squared})

powers.head()
```

```python
alt.Chart(powers).mark_line().encode(
    x= alt.X("X" , title= "X-axis"),
    y= alt.Y("Squared" , title= "Y-axis")
).properties(title= "Y-axis vs X-asis by powers")
```

Bar Charts and Histograms

```python
Dataframe = pd.DataFrame({'X': [1,2,3] , 'Y' : ["A","B","C"]})
print(Dataframe)
```

```python
alt.Chart(Dataframe).mark_bar().encode(
    x= "Y",
    y= "X"
)
```
